```python
import numpy as np
import pandas as pd

# Tampilkan semua kolom
pd.set_option('display.max_columns', None)
```

```python
base_dir = "C:\Purwadhika\Dataset\\"
path = base_dir + "Airbnb_Bangkok.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df
```

```python
df["name"].fillna("Unnamed Listing", inplace=True)
df["reviews_per_month"].fillna(0, inplace=True)
df["last_review"] = pd.to_datetime(df["last_review"], errors='coerce')
```

```python
df[df['name'] == 'Unnamed Listing']
```

```python
df[df['reviews_per_month'] == 0]
```

```python
# Mengecek jumlah duplikat pada tiap kolom
duplicate_counts = {col: df.duplicated(subset=[col]).sum() for col in df.columns}
duplicate_counts
```

```python
df.duplicated().sum()
```

```python
df[['price', 'minimum_nights', 'reviews_per_month']].describe()
```

```python
df = df[df['price'] > 0]
df = df[df['minimum_nights'] <= 365]  # Misal, 1 tahun dianggap maksimal wajar
```

```python
df.info()
```

```python
df['price'].min()
```

```python
df['minimum_nights'].max()
```

```python
# Apakah semua lokasi berada di Bangkok?
df[['latitude', 'longitude']].describe()
```

```python
duplicate_subset = ["name", "host_id", "latitude", "longitude"]

duplicates = df[df.duplicated(subset=duplicate_subset, keep=False)]

#hasilnya
print(f"Jumlah duplikat berdasarkan kombinasi {duplicate_subset}: {len(duplicates)}")
duplicates.sort_values(by=duplicate_subset).head(10)
```

```python
df["room_type"] = df["room_type"].str.strip().str.lower()
```

```python
df['price'].max()
```

```python
df['price'].min()
```

```python
price_99_5 = df["price"].quantile(0.995)
df["is_price_outlier"] = df["price"] > price_99_5
```

```python
df['price'].max()
```

```python
df["neighbourhood"] = df["neighbourhood"].str.strip().str.title()
```

```python
df['neighbourhood']
```

```python
df.rename(columns={'neighbourhood': 'district'}, inplace=True)
```

```python
df.to_csv(base_dir + "Airbnb_Bangkok_cleaned.csv", index=False)
```
