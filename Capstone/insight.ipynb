#### 2. Anlisis lokasi properti yang populer

```python
popular_neighbourhoods = df['neighbourhood'].value_counts().reset_index()
popular_neighbourhoods.columns = ['neighbourhood', 'number_of_listings']
popular_neighbourhoods.head(10)  # 10 lokasi terpopuler
```

Berdasarkan data analisis properti yang populer didapatkanlah rata-rata harga berdasarkan daerah dengan banyaknya jumlah properti seperti dibawah ini


```python
# Membuat kolom baru 'avg_price' pada popular_neighbourhoods berdasarkan rata-rata price di df
avg_price_per_neighbourhood = df.groupby('neighbourhood')['price'].mean().reset_index()
avg_price_per_neighbourhood.columns = ['neighbourhood', 'avg_price']

# Merge dengan popular_neighbourhoods
popular_neighbourhoods = popular_neighbourhoods.merge(avg_price_per_neighbourhood, on='neighbourhood')
popular_neighbourhoods.head(10)
```

```python
# Cek missing Value untuk semua kolom
missing_value_percentage = df.isna().sum() / len(df) * 100

missing_summary = pd.DataFrame({
                                'Missing_count': df.isna().sum(),
                                'Missing_Percentage': missing_value_percentage.round(2)
})
```
